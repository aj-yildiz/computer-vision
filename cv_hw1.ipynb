```python
# Import required libraries
import cv2 as cv
from matplotlib import pyplot as plt

# Step 1: Load the Image
image_path = 'massimo_photo.jpg'  # Replace this with the name of your uploaded file
img = cv.imread(image_path, cv.IMREAD_COLOR)  # Load the image in color mode
assert img is not None, "Image could not be loaded. Check the file path or format."

# Verify that the image is loaded correctly
print(f"Original Image Dimensions: {img.shape}")

# Step 2: Display the Original Image with Correct Colors
# Convert BGR to RGB for matplotlib (since OpenCV loads images in BGR by default)
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)

# Display the original image
plt.imshow(img_rgb)
plt.title('Original Image (RGB)')
plt.axis('off')  # Turn off axis
plt.show()

# Step 3: Resize the Image
# Maintain aspect ratio, with height = 300 pixels
height, width = img.shape[:2]
new_height = 300  # Set the new height to 300 pixels
aspect_ratio = width / height  # Calculate the aspect ratio
new_width = int(new_height * aspect_ratio)  # Calculate the new width based on the aspect ratio

# Resize the image using INTER_AREA for good results when downscaling
resized_img = cv.resize(img, (new_width, new_height), interpolation=cv.INTER_AREA)
print(f"Resized Image Dimensions: {resized_img.shape}")

# Step 4: Display the Resized Image with Correct Colors
# Convert BGR to RGB for matplotlib
resized_img_rgb = cv.cvtColor(resized_img, cv.COLOR_BGR2RGB)

# Display the resized image
plt.imshow(resized_img_rgb)
plt.title('Resized Image (RGB)')
plt.axis('off')  # Turn off axis
plt.show()

# Step 5: Rotate the Image by 180 Degrees
# Rotate the image upside down (180 degrees)
rotated_img = cv.rotate(resized_img, cv.ROTATE_180)

# Convert BGR to RGB for matplotlib
rotated_img_rgb = cv.cvtColor(rotated_img, cv.COLOR_BGR2RGB)
print(f"Rotated Image Dimensions:", {rotated_img.shape})

# Step 6: Display the Rotated Image
plt.imshow(rotated_img_rgb)
plt.title('Rotated Image (180 degrees)')
plt.axis('off')  # Turn off axis
plt.show()

# Step 7: Save the Resized Image
# Save the resized image to verify correctness
output_path = 'resized_image.jpg'
is_saved = cv.imwrite(output_path, resized_img)
print(f"Resized Image saved successfully: {is_saved}")

# Save the rotated image
rotated_output_path = 'rotated_image.jpg'
rotated_is_saved = cv.imwrite(rotated_output_path, rotated_img)
print(f"Rotated Image saved successfully: {rotated_is_saved}")
```
